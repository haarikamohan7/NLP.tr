```python
from wordcloud import WordCloud,STOPWORDS,ImageColorGenerator
import pandas as pd
import matplotlib.pyplot as plt
dataset = "http://data.insideairbnb.com/canada/on/toronto/2023-03-09/data/listings.csv.gz"
df = pd.read_csv(dataset)

descriptions =  df.dropna(subset=["description"], axis=0) ['description']
all_descriptions = " ".join(s for s in descriptions)
print("no_of_words:{}".format(len(all_descriptions)))
wordcloud = WordCloud(width = 800, height = 800,
                background_color ='white',
                stopwords = set(STOPWORDS),
                min_font_size = 10).generate(all_descriptions)

plt.figure(figsize=(10,10))
plt.imshow(wordcloud,interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
import re
def clean_text(text):
  text = re.sub(r'RT', '', text)
  text = re.sub(r'&amp', '', text)
  text = re.sub(r'!.;:,#@-', '', text)
  text= text.lower()
  text = re.sub(r'\d', '', text)
  text = re.sub(r'%', '', text)
  text = re.sub(r'[<>\\.?\\https:\\//]*','',text)
  return text

```

```python
import spacy
nlp = spacy.load("en_core_web_sm")
token=[]
pos=[]
for sent in nlp.pipe(df['sentence']):
  if sent.has_annotation('DEP'):
    token.append([word.text for word in sent])
    pos.append([word.pos_ for word in sent])
```

```python
df['token']= token
df['pos'] = pos
```

```python
df ['noun'] = df.apply(lambda x: x['pos'].count('NOUN'), axis=1)
df ['verb'] = df.apply(lambda x: x['pos'].count("VERB"), axis=1)
df ['adj'] = df.apply(lambda x: x['pos'].count('ADJ'), axis=1)
df['punct'] = df.apply(lambda x: x['pos'].count('PUNCT'), axis=1)
df
```
